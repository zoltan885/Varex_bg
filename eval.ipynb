```python
import numpy as np
import fabio
import matplotlib.pyplot as plt
```

```python
imgs = {
    7: 'snap_V3_00243.dark.cbf',
    6: 'snap_V3_00245.dark.cbf',
    5: 'snap_V3_00247.dark.cbf',
    4: 'snap_V3_00249.dark.cbf',
    3: 'snap_V3_00251.dark.cbf',
    2: 'snap_V3_00253.dark.cbf',
       }
data = {k: fabio.open(v).data for k,v in imgs.items()}
mean = np.array([np.mean(d) for d in data.values()])
std = np.array([np.std(d) for d in data.values()])
```

```python
for i,(g,d) in enumerate(data.items()):
    plt.imshow(d, vmin=0, vmax=12000)
    plt.title('gain %i: mean %.1f std %.2f' % (g, mean[i], std[i]))
    plt.show()
```

```python
plt.plot(data.keys(), mean)
plt.xlabel('gain')
plt.ylabel('mean')
plt.fill_between(data.keys(), mean-std, mean+std, facecolor='b', alpha=0.2)
plt.show()
```
